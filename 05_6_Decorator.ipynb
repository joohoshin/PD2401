Decorator를 통해 기존 함수를 변경하지 않고 추가적인 작업을 할수 있게 만들 수 있다. 


```python
def print_name(name):
  print(name)
```

```python
print_name('j2')
```

```python
# 출력 시간을 추가해보자

import datetime

def add_time(func):
    def wrapper(*args, **kwargs):        
        print(func(*args, **kwargs))
        print(datetime.datetime.now())
    return wrapper
```

```python
@add_time
def print_name_time(name):
  return name
```

```python
print_name_time('ttt')
```
