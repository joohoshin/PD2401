폴더에 있는 여러 개 이미지를 불러오겠습니다

```python
import os  
from PIL import Image
```

```python
# 폴더 경로 지정
path = '/content/drive/MyDrive/Teaching/DL 202301/imgs'
```

```python
files = os.listdir(path)  # 폴더에 있는 파일 불러오기
```

```python
files
```

```python
imgs = []

for file in files:  # 파일명 순회
  file_path = os.path.join(path, file) # 경로 + 파일명
  img = Image.open(file_path) # 이미지 열기
  imgs.append(img) # 리스트에 추가
```
