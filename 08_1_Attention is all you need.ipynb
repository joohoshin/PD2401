* RNN, LSTM, GRU 등은 h_t-1을 h_t에서 참고해야해서 병렬 처리가 힘들다
* Transformer 에서는 Recurrence를 없애고, 어텐션에만 사용해서 병렬처리가 가능하도록 했다
