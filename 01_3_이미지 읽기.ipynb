PIL 라이브러리를 사용하여 불러들여봅시다.
PIL은 파이썬 기본 이미지 라이브러리입니다

https://pillow.readthedocs.io/en/stable/

```python
from PIL import Image

# 경로
file = "/content/drive/MyDrive/Teaching/PD 2401/DALL·E 2024-03-05 15.42.43 - A lively and colorful illustration featuring 2 to 3 young people, radiating positivity and youthfulness. These individuals are depicted in a setting t.webp"

# 파일 열기
img = Image.open(file)
img.show()
```

numpy 배열로 변환해봅시다

```python
import numpy as np
img_np = np.array(img)
```

```python
img_np.shape
```

PIL 이미지는 기본적으로 RGB 색상 모드로 열리며, numpy 배열은 높이 x 너비 x 채널 형태입니다.

따라서, 위 코드에서 변환된 numpy 배열도 높이 x 너비 x 채널 형태입니다.

matplotlib 라이브러리(차트)에서 numpy 배열과 pil 이미지를 출력할 수 있습니다.

```python
import matplotlib.pyplot as plt

plt.imshow(img_np) # numpy 배열 출력
```

```python
plt.imshow(img)  # pil image 출력
```

OpenCV 라이브러리를 사용해서 이미지를 열수 있습니다.

OpenCV는 동영상, 실시간 영상 등 다양한 형태의 이미지를 불러오고 처리할 수 있습니다.

https://pypi.org/project/opencv-python/

```python
import cv2

# 이미지 불러오기
img_cv = cv2.imread(file)

# 이미지 보여주기
plt.imshow(img_cv[:,:,::-1]) # BGR -> RGB로 변경
plt.show()
```

OpenCV에서 불러들이면, numpy 배열로 저장됩니다.

```python
type(img_cv)
```
