특정 조건으로 필터링 가능

```python
df['country_name'].str.contains("korea", case=False)
```

df[] 괄호 안에 True, False를 넣게되면 True인 경우만 선택됩니다. 이를 활용해서 위의 결과가 참인 경우만 출력해봅시다

```python
kor=df[df['country_name'].str.contains("korea", case=False)]
kor
```

```python
south_kor = kor.query('country_code=="KOR"')
```

```python
south_kor.dtypes
```

```python
import matplotlib.pyplot as plt

plt.plot(south_kor['year'], south_kor['value'])
```
