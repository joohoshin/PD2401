데이터를 tensor로 변경합시다
이때 float32로 타입을 변경합니다.

torch에서 모델 만들때 파라미터 데이터 타입이 기본적으로 float32를 사용하기 때문에 지정 안할 시 float64로 변환되면 오류 발생합니다

```python
import torch
from torch import nn
```

```python
x = torch.from_numpy(x).float()
y = torch.from_numpy(y).float()
```

2개의 레이어로 이루어진 모델을 만들어보겠습니다.

nn.Sequential()을 활용해서 구성해보겠습니다.

13개의 Input이 있고, 최종적으로 1개의 output이 있습니다.

```python
model = nn.Sequential(
    nn.Linear(13, 64),
    nn.ReLU(),
    nn.Linear(64, 1)
)
```

```python
model
```

```python
# 모델 파라미터의 데이터 타입 확인 코드입니다

for name, param in model.named_parameters():
    print(name, param.dtype)
```

```python
model(x[:10]) # 모델이 작동하는지 확인해봅니다
```

```python
model(x[:10]).shape  #모델의 output은 2차원으로 나옴
```

loss function과 optimizer를 지정합니다.

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

모델 학습을 해봅시다.
이때 데이터셋을 batch 단위로 나누어서 학습을 해보겠습니다.

```python
epochs = 1000
batch_size = 10
```

```python
for epoch in range(epochs):
  running_loss = 0 # 전체 데이터셋에 대한 loss 계산

  # 데이터를 batch로 나누어서 학습
  for i in range(0, x.shape[0], batch_size): #range(start, end, step), end는 미포함
    inputs = x[i:i+batch_size,:]
    labels = y[i:i+batch_size]

    optimizer.zero_grad() # gradient 초기화
    outputs = model(inputs) # 모델값 예측
    #print(outputs)
    loss = loss_fn(outputs, labels.unsqueeze(dim=1)) # 에러 계산, unsqueeze(): 모델의 output과 원래 데이터 shape 맞추기
    loss.backward() # gradient 계산
    optimizer.step() # gradient 업데이트

    running_loss += loss.item() * batch_size

  if (epoch+1)%10==0:
    print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, epochs, running_loss / x.shape[0]))

```
